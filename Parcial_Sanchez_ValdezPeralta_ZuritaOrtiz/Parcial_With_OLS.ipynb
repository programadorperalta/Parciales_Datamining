```python
### Matrix of correlations
def matrix_correlation(X,Y): 
    df_test = pd.concat((Y,X), axis=1)
    correlations = df_test.corr()
    return correlations
```

```python
matrix_correlation(x,y).style.background_gradient(cmap='coolwarm')
```

```python
## Search features with low correlation.
def search_correlation(x,y):
    elements = matrix_correlation(x,y)[0:1].abs()
    columns = x.columns
    array_low_correlation = []
    array_high_correlation = []
    for col in columns: 
        if((elements[col].values>0) & (elements[col].values<=0.10)):
            print(f"Correlación baja inexistente con {y.columns[0]}: ",col)
            array_low_correlation.append(col)
        if((elements[col].values>=0.50) & (elements[col].values<1)):
            print(f"Correlación alta inexistente con {y.columns[0]}: ",col)
            array_high_correlation.append(col)
    return array_low_correlation,array_high_correlation
```

```python
array_low,array_high = search_correlation(x,y)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
import pylab

from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
def ols_regression(X,Y,size): 
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=size, random_state=42)
    ml = sm.OLS(y_train, sm.add_constant( x_train))
    ml_results = ml.fit()
    return x_train, y_train,x_test, y_test,ml
```
