```python
### Matrix of correlations
def matrix_correlation(X,Y): 
    df_test = pd.concat((Y,X), axis=1)
    correlations = df_test.corr()
    return correlations
```

```python
matrix_correlation(x,y).style.background_gradient(cmap='coolwarm')
```

```python
## Search features with low correlation.
def search_correlation(x,y):
    elements = matrix_correlation(x,y)[0:1].abs()
    columns = x.columns
    array_low_correlation = []
    array_high_correlation = []
    for col in columns: 
        if((elements[col].values>0) & (elements[col].values<=0.10)):
            print(f"Correlacin baja inexistente con {y.columns[0]}: ",col)
            array_low_correlation.append(col)
        if((elements[col].values>=0.50) & (elements[col].values<1)):
            print(f"Correlacin alta inexistente con {y.columns[0]}: ",col)
            array_high_correlation.append(col)
    return array_low_correlation,array_high_correlation
```

```python
array_low,array_high = search_correlation(x,y)
```

```python
def convert(type_normalization,df):
    rescaledX = type_normalization.fit_transform(df) 
    dt_result = pd.DataFrame(rescaledX,index=df.index,columns=df.columns)
    return dt_result
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler,Normalizer

def normalization(df):
    dt_result = any
    array = df.to_numpy()
    result = np.where(array<0)
    if(len(result)>2):
        print('Se escalar con el metodo StandarScaler()')
        dt_result = convert(StandarScaler(),df)
        normalization = True
    else:
        print('Se escalar con el metodo MinMaxScaler()')
        dt_result = convert(MinMaxScaler(),df)
        normalization = True
    return dt_result
```
