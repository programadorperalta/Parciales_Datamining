# Análisis exploratorio de datos

¿Está listo para hacer el pago inicial de su primera casa? ¿O busca aprovechar el capital de la casa que tiene? Para respaldar las necesidades de una variedad de decisiones financieras, el Banco Santander  ofrece una mano amiga a sus clientes a través de recomendaciones personalizadas de productos.

Bajo su sistema actual, un pequeño número de clientes de Santander recibe muchas recomendaciones, mientras que muchos otros rara vez ven alguna, lo que resulta en una experiencia de cliente desigual. En su segunda competencia, Santander desafía a Kagglers a predecir qué productos utilizarán sus clientes actuales en el próximo mes en función de su comportamiento anterior y el de clientes similares. 

En esta competencia, se le proporciona 1,5 años de datos de comportamiento de los clientes del banco Santander para predecir qué nuevos productos comprarán los clientes. Los datos comienzan en 2015-01-28 y tienen registros mensuales de productos que tiene un cliente, como "tarjeta de crédito", "cuenta de ahorros", etc.<h5>Predecirá qué productos adicionales obtendrá un cliente en el último mes 2016- 28-06, además de lo que ya tienen el 28-05-2016.</h5> 
<h4>Estos productos son las columnas denominadas: ind_(xyz)_ult1, que son las columnas #25 - #48 en los datos de entrenamiento</h4> 
<h5>Pronosticará qué comprará un cliente además de lo que ya tenía el 28-05-2016</h5>
Los conjuntos de prueba y entrenamiento se dividen por tiempo, y los conjuntos de tablas de clasificación públicas y privadas se dividen aleatoriamente.

#### Descripcion de columnas 
1. fecha_dato: Comienzo de la toma de datos. 
2. ncodpers: codigo del cliente 
3. ind_empleado: indice de empleados A activo, B ex empleado, F filial, N no empleado, P pasivo
4. pais_residencia: Pais de residencia del cliente
5. sexo: sexo del cliente
6. age: edad del cliente
7. fecha_alta: La fecha en que el cliente pasó a ser el primer titular de un contrato en el banco
8. ind_nuevo: Índice de nuevos clientes. 1 si el cliente se registró en los últimos 6 meses.
9. antiguedad: Antigüedad del cliente (en meses)
10. indrel: 1 (Primero/Principal), 99 (Cliente principal durante el mes pero no al final del mes)
11. ult_fec_cli_1t: Última fecha como cliente principal (si no es a fin de mes)
12. indrel_1mes: Tipo de cliente al comienzo del mes, 1 (primer cliente/principal), 2 (copropietario), P (potencial), 3 (antiguo principal), 4 (antiguo copropietario)
13. tiprel_1mes: Tipo de relación del cliente al inicio del mes, A (activo), I (inactivo), P (antiguo cliente), R (Potencial)
14. indresi: Índice de residencia (S (Sí) o N (No) si el país de residencia es el mismo que el del banco)
15. indext: Índice de extranjería (S (Sí) o N (No) si el país de nacimiento del cliente es diferente al país del banco)
16. conyuemp: Índice de cónyuge. 1 si el cliente es cónyuge de un empleado
17. canal_entrada: canal utilizado por el cliente para unirse
18. indfall: Índice de fallecidos. N/D
19. tipodom: Tipo de dirección. 1, dirección principal
20. cod_prov: Código de provincia (dirección del cliente)
21. nomprov: Nombre de la provincia
22. ind_actividad_cliente: Índice de actividad (1, cliente activo; 0, cliente inactivo)
23. renta: Ingreso bruto del hogar
24. segmento: segmentación: 01 - VIP, 02 - Particulares 03 - graduado universitario
25. ind_ahor_fin_ult1: Cuenta de ahorro 
26. ind_aval_fin_ult1: Garantías
27. ind_cco_fin_ult1: Cuentas actuales
28. ind_cder_fin_ult1: Cuenta derivada
29. ind_cno_fin_ult1: Cuenta de Nómina
30. ind_ctju_fin_ult1: Cuenta Junior
31. ind_ctma_fin_ult1: Cuenta más particular
32. ind_ctop_fin_ult1: cuenta particular
33. ind_ctpp_fin_ult1: Cuenta Plus en particular
34. ind_deco_fin_ult1: Depósitos a corto plazo
35. ind_deme_fin_ult1: Depósitos a mediano plazo
36. ind_dela_fin_ult1: Depósitos a largo plazo
37. ind_ecue_fin_ult1: cuenta electronica
38. ind_fond_fin_ult1: Fondos
39. ind_hip_fin_ult1: Hipoteca
40. ind_plan_fin_ult1: Pensiones

43. ind_tjcr_fin_ult1: Tarjeta de crédito

### Obtención de recomendación para un usuario random (que exista en el dataset)

```python
from numpy import dot
from numpy.linalg import norm
```

```python
def get_label_name(label, le=le):
    return data_desc[data_desc['Column Name'] == le.inverse_transform([label])[0]]['Description'].values[0]
```

```python
def cosine_sim(X,Y):
    return dot(X,Y) / (norm(X)*norm(Y))
```

```python
def get_sim_user_recommendation(uid, top_n, X):
    # Buscar un usuario especifico 
    user_specified = X.loc[uid]
    
    # Calcule la similitud con todos y cada uno de los usuarios.
    res = X.progress_apply(lambda user: cosine_sim(user_specified, user), axis=1)
    
    # Convertir a Dataframe
    res = res.to_frame(name='sim_score')
    
    # Borrar el índice y convertirlo en una columna.
    res.reset_index(inplace=True)
    
    # Unir los datos de usuario y la tabla res en ncodpers
    res = pd.merge(left= user_data[['ncodpers','servicio_seleccionado']], 
                   right = res, 
                   on='ncodpers')
    
    # Obtener la fila más similar de cada categoría de servicio
    res = res[~res['servicio_seleccionado'].duplicated(keep='first')]
    
    # Ordenar los resultados
    res.sort_values(by='sim_score', ascending=False, inplace=True)
    
    # Agregar una columna de service_seleccionado_name
    res['service_seleccionado_name'] = res['servicio_seleccionado'].progress_apply(lambda label: get_label_name(label, label_encoder))
    
    # Borra el índice y convertirlo en una columna.
    res.reset_index(drop=True, inplace=True)
    
    # Retornar las predicciones 
    return res
```

### Recomendación

```python
def recomendacion_de_producto(dataframe):
    recomendations_high = []
    recomendations_low = []
    
    cols = data_columns
    for value in range(len(dataframe)):
        for col in cols:
            ## Eligiendo el criterio del 30%, decidimos que si la probabilidad es mayor o igual al 30% se recomendarà el o los productos. 
            if(dataframe[col][value]*100 >= 20):
                recomendations_high.append(col)
            else:
                recomendations_low.append(col)
                
    return recomendations_high,recomendations_low
```
